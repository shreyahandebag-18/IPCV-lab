```python
# To implement various image transformation techniques for image enhancement such as:
# • Image Negative
# • Contrast Stretching
# • Bit Plane Slicing
# • Gray Level Slicing
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
image_path =('/content/7-colorful-photography.jpg')
```

```python
img = cv2.imread(image_path)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Original Image (RGB)")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(img_gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
#• Image Negative
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('7-colorful-photography.jpg', 0)
negative = 255 - img

plt.subplot(1,2,1), plt.title("Original"), plt.imshow(img)
plt.subplot(1,2,2), plt.title("Negative"), plt.imshow(negative, cmap='gray')
plt.show()
```

```python
#• Contrast Stretching
```

```python
import numpy as np

r_min = img.min()
r_max = img.max()

contrast_stretched = ((img - r_min) / (r_max - r_min)) * 255
contrast_stretched = contrast_stretched.astype(np.uint8)

plt.imshow(contrast_stretched, cmap='gray')
plt.title("Contrast Stretched Image")
plt.show()
```

```python
#• Bit Plane Slicing
```

```python
bit_planes = []

for i in range(8):
    bit_plane = (img >> i) & 1
    bit_planes.append(bit_plane * 255)

plt.figure(figsize=(10,5))
for i in range(8):
    plt.subplot(2,4,i+1)
    plt.title(f'Bit Plane {i}')
    plt.imshow(bit_planes[i], cmap='gray')
    plt.axis('off')

plt.show()
```

```python
#• Gray Level Slicing
```

```python
lower = 100
upper = 180

sliced = img.copy()
sliced[(img < lower) | (img > upper)] = 0

plt.imshow(sliced, cmap='gray')
plt.title("Gray Level Sliced Image")
plt.show()
```
